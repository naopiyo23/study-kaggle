```python
plt.figure(figsize=(18,8))
plt.subplot(1,2,1)
sns.countplot(train["Survived"])
plt.subplot(1,2,2)
train["Survived"].value_counts().plot.pie(explode=[0,0.1],autopct='%1.1f%%',shadow=True)
```

```python
plt.figure(figsize=(18,5))
plt.subplot(1,2,1)
sns.countplot(train["Pclass"])
plt.subplot(1,2,2)
sns.countplot(x='Pclass', hue='Survived', data=train)
```

```python
pd.crosstab(train["Pclass"], train["Survived"]).plot(kind='bar',stacked=True, figsize=(9,5))
```

```python
plt.figure(figsize=(18,5))
plt.subplot(1,2,1)
sns.countplot(train["Sex"])
plt.subplot(1,2,2)
sns.countplot(x='Sex', hue='Survived', data=train)
```

```python
pd.crosstab(train["Sex"], train["Survived"]).plot(kind='bar',stacked=True, figsize=(9,5))
```

```python
plt.figure(figsize=(18,5))
plt.subplot(1,2,1)
sns.distplot(train["Age"].dropna(), bins=70)
plt.subplot(1,2,2)
sns.violinplot(x='Survived',y='Age',data=train)
```

```python
plt.figure(figsize = (18, 5))

sns.kdeplot(train.loc[train['Survived'] == 0, 'Age'], label = 'Survived == 0')
sns.kdeplot(train.loc[train['Survived'] == 1, 'Age'], label = 'Survived == 1')

plt.xlabel('Age (years)'); plt.ylabel('Density'); plt.title('Distribution of Ages');
```
